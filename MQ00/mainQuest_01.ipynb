```python
# [메인퀘스트 1번]

import random                                        # 계좌번호를 생성하기 위해 랜덤 모듈을 불러온다


class Account:

    count_account_num = 0

    def __init__(self, who, balance):                           # 인스턴스 변수 선언
        self.deposit_cnt = 0                                    # 계설된 총 계좌 수

        self.bank = 'SC은행'
        self.account_num = ''                                   # 부여될 계좌번호를 담을 변수

        # 계좌 개설 시 입력 받을 정보 - 예금주, 초기잔액
        self.who = who
        self.balance = int(balance)
        self.history_deposit = []
        self.history_withdraw = []

        # 인스턴스가 생성될 때마다 총 개설 계좌수의 수를 1씩 늘려주도록 한다
        Account.count_account_num += 1

        # [[[[[ 계좌번호 부여를 위한 코드 : random 모듈을 사용하여 사용자마다 랜덤하게 생성된다 ]]]]]

        self.account_num_list = []                              # 계좌번호의 숫자들을 담을 리스트 변수

        # 계좌번호 자리 수만큼 random 모듈을 사용
        for i in range(11):
            x = random.randrange(0, 9)
            # 생성된 난수를 self.account_num_list에 붙여준다
            self.account_num_list.append(x)

        # 리스트에 담긴 난수 사이에 -를 넣어준다. 특정 인덱스에 문자 삽입하는 방법을 이용한다
        self.account_num_list.insert(3, '-')
        self.account_num_list.insert(6, '-')

        # 계좌번호 인스턴스 변수에 위 리스트를 풀어서 담아줌
        for i in range(len(self.account_num_list)):
            self.account_num += str(self.account_num_list[i]) + ''

      # [[[[[ get_account_num() 메서드 : Account 클래스로부터 생성된 계좌의 개수를 출력한다 ]]]]]
    def get_account_num(self):
        print(self.account_num)

    # [[[[[ deposit 메서드 : Account 클래스에 입금을 위한 메서드. 입금은 최소 1원 이상만 가능 ]]]]]

    def deposit(self, howmuch):                                   # 입금할 금액을 파라미터로 입력받음
        if howmuch < 1:                                           # 입금금액이 1보다 작을 경우 입금불가 처리
            print('1원 이상만 입금 가능')
        else:
            # 1보다 클 경우 입금 가능. balace변수에 누적해 담아준다
            self.balance += howmuch
            self.deposit_cnt += 1
            a = (f'{howmuch}원 입금되었습니다\n잔액 : {self.balance}')

            if self.deposit_cnt % 5 == 0:                        # 입금회차가 5의 배수일 경우 이자를 지급한다
                self.balance += self.balance * 0.01
                print(self.deposit_cnt)
                print(f'이자 지급회차입니다\n잔액 : {self.balance}')

        # 입금 내역을 저장하는 리스트 변수에 입금 기록 문자열을 붙인다
        self.history_deposit.append(a)

    # [[[[[ withdraw : 계좌에서 출금하는 메서드. 잔고 이상의 금액은 출금 불가 ]]]]]

    def withdraw(self, howmuch):                                  # 출금할 금액을 파라미터로 입력받는다
        if howmuch > self.balance:                                # 출금할 금액이 잔고 이상일 경우 지급 거절
            print('출금 가능 금액 초과')
        else:                                                      # 출금금액이 잔고 이하일 경우 balance에서 출금금액만큼 차감하여 저장
            self.balance -= howmuch
            a = f'{howmuch} 원이 인출되었습니다\n잔액 : {self.balance}'

        # 출금 내역을 저장하는 리스트 변수에 출금 기록 문자열을 붙인다
            self.history_withdraw.append(a)


# [[[[[ display_info() : Account 인스턴스에 저장된 정보를 출력하는 메서드. (잔고는 세자리마다 쉼표 출력) ]]]]]

# 출력 순서 맞추어 출력. 잔고의 경우 format 메서드 사용해 세 자리수 마다 쉼표 찍어주도록 한다


    def display_info(self):
        account_info = ['은행이름:', self.bank, '예금주:', self.who, '계좌번호:', self.account_num, '잔고: ', format(
            self.balance, ',')]
        print(account_info)
        return account_info


# 입금내역 출력
# 입금 시마다 기록을 저장했던 리스트 변수 self.history_deposit 값을 꺼내와 그 개수만큼 출력해준다

    def deposit_history(self):
        print('='*30)
        print('          입금내역           ')
        print('='*30)
        for i in self.history_deposit:
            print(i)


# 출금 내역 출력 - 출금 시마다 기록을 저장했던 리스트 변수 self.history_withdraw 값을 꺼내와 그 개수만큼 출력해준다

    def withdraw_history(self):
        print('='*30)
        print('          출금내역           ')
        print('='*30)
        for i in self.history_withdraw:
            print(i)


# 생성될 Account 인스턴스들을 저장할 리스트 변수
ls_account = []

acc_1 = Account('kim', 100000)                               # 4개의 인스턴스를 생성하고
acc_2 = Account('park', 250600)
acc_3 = Account('choi', 50000000000)
acc_4 = Account('lee', 30000000)
# ls_account 변수에 차례대로 붙여나간다
ls_account.append(acc_1.display_info())
ls_account.append(acc_2.display_info())
ls_account.append(acc_3.display_info())
ls_account.append(acc_4.display_info())


# 현재 개설 중인 계좌를 확인해본다
print('='*20, '개설된 계좌 정보', '='*20, '\n', ls_account)


# 잔고가 100만원 이상인 고객 정보 출력하기
print('='*20, '잔액 100만원 이상 고객 목록', '='*20)
# 아래의 반복문 실행 횟수를 저장할 변수
cnt = 0

for i in ls_account:
    # 해당 인덱스 계좌의 금액을 찾아 백만원이 넘으면 출력하도록 한다.
    if int(ls_account[cnt][7].replace(',', '')) >= 1000000:
        print(i)
    cnt += 1


acc_4.deposit(30000)
acc_4.deposit(60000)
acc_4.deposit_history()                                   # 입금 기록이 출력된다
acc_4.withdraw(60000)
acc_4.withdraw(60000)
acc_4.withdraw_history()                                   # 출금 기록이 출력된다
```
