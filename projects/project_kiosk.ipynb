```python
class Square:
    # 인스턴스 속성 - 인스턴스 생성 시 부여되는 속성을 지정한다 -> init으로!
    def __init__(self):
        self.shape = int(
            input('어떤 사각형의 넓이를 구할까요?\n1. 직사각형\n2. 평행사변형\n3. 사다리꼴 \n'))

        if self.shape == 1:
            print('직사각형 함수는 : rect()')

        elif self.shape == 2:
            print('평행사변형 함수는 : par()')
        elif self.shape == 3:
            print('사다리꼴 함수는 : trape()')
        else:
            print('1, 2, 3 중에서만 고르세요')

    def rect(self):
        # 가로, 세로 값 받아 / 쪼개서 / int로 변환 / 곱
        width, height = map(
            int, (input('가로, 세로의 값을 띄어서 입력하세요 예) 2 3 : ').split()))
        area = width * height
        result = '직사각형의 넓이는 ' + str(area) + ' 입니다'
        return result
        print(width, height)

    def par(self):
        bottom, height = map(int, input(
            '평행사변형의 밑변, 높이를 띄어서 써주세요. ex) 2 3 : ').split())
        area = bottom * height
        result = '평행사변형의 넓이는 ' + str(area) + ' 입니다'
        return result

    def trape(self):
        top, bottom,  height = map(int, input(
            '사다리꼴의 윗변, 밑변, 높이 순으로 띄어서 써주세요. 예) 2 3 5 : ').split())

        area = ((top + bottom) * height) / 2
        result = '사다리꼴의 넓이는 '+str(area) + ' 입니다'
        return result


square1 = Square()
print(square1.rect())  # 직사각형의 넓이 구하기
print(square1.par())  # 평행사변형의 넓이 구하기
print(square1.trape())  # 사다리꼴의 넓이 구하기
```
